# Python 的特色
Python 是一門直譯器語言，需要在直譯器上運行。相比編譯後運行的語言，運行速度上會處於劣勢。好處是跨平台簡單，一處開發多處運行。

Python 是物件導向的語言，包括函數、字串、數字甚至類型在內都是物件。

Python 的第三方庫豐富，因此開發極其方便，程式碼量極少。
